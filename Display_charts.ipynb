```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_red = pd.read_csv('winequality-red.csv', sep=";")
df_white = pd.read_csv('winequality-white.csv', sep=";")
```

```python
# Create a list of attributes to plot
attributes = df_red.columns[:-1]

# Create a figure with subplots
fig, axs = plt.subplots(nrows=len(attributes), ncols=2, figsize=(12, 24))

# Iterate over attributes and plot them against quality
for i, attr in enumerate(attributes):
    axs[i, 0].scatter(df_red[attr], df_red['quality'], alpha=0.5)
    axs[i, 0].set_title(f"{attr} (Red Wine)")
    axs[i, 0].set_xlabel(attr)
    axs[i, 0].set_ylabel('Quality')
    
    axs[i, 1].scatter(df_white[attr], df_white['quality'], alpha=0.5)
    axs[i, 1].set_title(f"{attr} (White Wine)")
    axs[i, 1].set_xlabel(attr)
    axs[i, 1].set_ylabel('Quality')

# Adjust spacing and display the figure
fig.tight_layout()
plt.show()

# Create a box plot for each attribute against quality
for column in df_red.columns:
    if column == 'quality':
        continue
    fig, ax = plt.subplots()
    ax.boxplot([df_red[column][df_red['quality'] == q] for q in range(3, 9)])
    ax.set_xlabel('Quality')
    ax.set_ylabel(column)
    ax.set_title(column)
    plt.show()
```
